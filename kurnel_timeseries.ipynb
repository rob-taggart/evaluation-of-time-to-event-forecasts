```python
obs = pd.read_csv('data/kurnell_timeseries_observations.csv')
fcst = pd.read_csv('data/kurnell_timeseries_forecasts.csv')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 4))


max_value = 25
text_yplacement = max_value - 0.5


ax.plot(
    [0, 18], [15, 15],
    color='#777777', linestyle='dashed'
)


ax.plot(
    np.arange(len(fcst)), fcst['windmag'],
    label='forecast', color='#d95f02', marker='o'
)

ax.plot(
    np.arange(len(obs))/60, obs['windmag'], 
    label='realization', color='#1b9e77'
)


ax.legend(loc='upper left')

ax.set_xticks([0, 6, 12, 18])
ax.set_xlabel("Hours since 3 January 2023 18:00 UTC")
ax.set_ylabel("Wind speed (knots)")

plt.show()
```
